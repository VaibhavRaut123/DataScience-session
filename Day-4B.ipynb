#### separate the input and output

```python
x = df[['disp','hp','wt']]
y= df['mpg']
```

```python
x
```

#### import the class

```python
from sklearn.linear_model import LinearRegression
```

### create the object

```python
reg = LinearRegression()
```

### Decision Tree Classification

### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
df # 0 means purchased 1 means not purchased
```

#### separate input and output variable

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
x;
```

```python
# how many values are present in each category of output is known using series function ==. Value_counts()
```

```python
y.value_counts()
```

#### Visualise the data

```python
plt.title('Purchased data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c=y) #see whether groups are present ==> c=y
```

### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

### train the model

```python
classifier.fit(x,y)
```

### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(classifier, fontsize=7, feature_names=['age', 'sal'], class_names=['y','n'], filled=True, rounded=True);
```

### predict on new data

```python
new = pd.DataFrame({
    'Age': [23,45,54,67],
    'EstimatedSalary':[4500,89000,23000,134000]
})
```

```python
new
```

```python
classifier.predict(new) #predict whether above four will purchase or not
```
