### LINE PLOT

```python
x = np.array([3,11,16,23,27,34,48,55,56,67])
y = np.array([13,23,34,49,47,61,69,78,84,89])
```

```python
z = np.array([87,77,75,71,54,33,21,23,19,10])
```

```python
plt.figure(figsize=(3,4)) # aspect ratio
plt.title('My first plot')

plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))
plt.yticks (range(0,101,10))
plt.plot(x,z)
# marker: o > < v ^ D *
plt.plot(x, y, marker = 'o', color = 'red', ls =':',lw = 3, mfc ='white' )
```

```python
plt.title('My first plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.plot(x, y, label='2022', color='blue')
plt.plot(x,z, label='2023', color='green')
plt.legend () # for showing label name
```

```python
plt.figure(figsize=(3,4))
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(1,2,1)
# 1 row, 2 column and plot 1
plt.plot(x,y)
plt.subplot(1,2,2)
# 1 row, 2 column and plot 2
plt.plot(x,z)
```
